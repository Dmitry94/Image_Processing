```python
class BinTree:
    def __init__(self, val):
        self.left = None
        self.right = None
        self.val = val
    
    def set_left(self, val):
        self.left = BinTree(val)
        
    def set_right(self, val):
        self.right = BinTree(val)
        
    @property
    def dist(self):
        return self.val[1] - self.val[0]
    
    def split(self, val):
        self.set_left((self.val[0], val))
        self.set_right((val, self.val[1]))
    
    def finish_branch(self):
        self.set_left(None)
        self.set_right(None)
        
    @property
    def height(self, h=0):
        lh = self.left and self.left.height(h+1) or -1
        rh = self.left and self.right.height(h+1) or -1
        return h + max(rh,lh)
    
    def satisfy(self,val):
        if self.val is None:
            return False
        return threshed(*self.val)(val)
    
    def decide(self, val, left=True):
        if self.left.satisfy(val):
            return self.left.decide(val, True)
        if self.right.satisfy(val):
            return self.right.decide(val, False)
        return left   
```

```python
def threshed(mn, mx):
    def crit(x):
        return mn < x and x < mx
    return crit

def calc_hist(mn, mx):
    crit = threshed(mn, mx)
    vals = img.mean(axis=2).flatten()
    vals = list(filter(crit, vals))
    hist, bins = np.histogram(vals, range(256))
    return hist

def otsu_thresh(mn, mx):
    hist = calc_hist(mn, mx)
    n = 256
    wt0 = list(scanl(map(np.int_, hist), op.add))
    wt1 = list(map(lambda x: wt0[-1] - x, wt0))
    mu0 = list(scanl(map(lambda p,i,w: i*p/w, hist, range(0,n), wt), op.add))
    mu1 = list(map(lambda x: mu0[-1] - x, mu0))
    sigma = list(map(lambda w0,w1,m0,m1: w0*w1*(m0-m1)**2, wt0, wt1, mu0, mu1))
    return (np.max(sigma), np.argmax(sigma))
```

```python
def hierarch(tree=None):
    if tree is None:
        tree = BinTree((0,255))
    (d,t) = otsu_thresh(*tree.val)
    if d < 1e10 or tree.dist < 50:
        tree.finish_branch()
        return tree
    tree.split(t)
    hierarch(tree.left)
    hierarch(tree.right)  
    return tree
```
